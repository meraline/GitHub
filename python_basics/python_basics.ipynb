## Задание 1

Даны 2 переменных, в которых хранятся строки произвольной длины: *phrase_1* и *phrase_2*.  
Напишите код, который проверяет какая из этих строк длиннее.

Примеры работы программы:  
1.  

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

```

Результат:  
`Фраза 1 длиннее фразы 2`
2.     

```
phrase_1 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:  
`Фраза 2 длиннее фразы 1`
3.     

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:  
`Фразы равной длины`

## Решение  1

````python
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

if len(phrase_1)>len(phrase_2):
    print('Фраза 1 длиннее фразы 2')
elif len(phrase_1)<len(phrase_2):
    print('Фраза 2 длиннее фразы 1')
else:
    print('Фразы равной длины')
````

## Задание 2

````python
# Козерог: с 20 января по 16 февраля 01.21 - 02.16
# Водолей: с 16 февраля по 11 марта 02.17 - 03.11
# Рыбы: с 11 марта по 18 апреля  03.12 - 04.18
# Овен: с 18 апреля по 13 мая 04.19 - 05.13
# Телец: с 13 мая по 21 июня 05.14 - 06.21
# Близнецы: с 21 июня по 20 июля 06.22 - 07.20
# Рак: с 20 июля по 10 августа 07.21 - 08.10
# Лев: с 10 августа по 16 сентября 08.11 - 09.16
# Дева: с 16 сентября по 30 октября 09.17 - 10.30
# Весы: с 30 октября по 23 ноября 10.31 - 11.23
# Скорпион: с 23 ноября по 29 ноября 11.24 - 11.29
# Змееносец: с 29 ноября по 17 декабря 11.30 - 12.17
# Стрелец: с 17 декабря по 20 января 12.18 - 01.20
    

day = int(input('Введите день: \n'))
if 0 < day < 32:
    month = input('Введите месяц: \n')
    if (month == ('январь' or 'января') and (day<=31 and day>=21))
    or (month == ('февраль' or 'февраля')and (day>=1 and day <=16)):
        print('Ваш знак зодиака: Козерог')
    elif (month == ('февраль' or 'февраля') and (day<=29 and day>=17))
    or (month == ('март' or 'марта') and (day>=1 and day <=11)):
        print('Ваш знак зодиака: Водолей') 
    elif (month == ('март' or 'марта') and (day<=31 and day>=12))
    or ( month == ('апреля' or 'апрель') and (day>=4 and day <=18)):
        print('Ваш знак зодиака: Рыбы')
    elif (month == ('апреля' or 'апрель') and (day<=30 and day>=19))
    or (month == ('мая' or 'май') and (day>=1 and day <=13)):
        print('Ваш знак зодиака: Овен') 
    elif (month == ('мая' or 'май') and (day<=31 and day>=14))
    or ( month == ('июня' or 'июнь') and (day>=1 and day <=21)):
        print('Ваш знак зодиака: Телец')
    elif (month == ('июня' or 'июнь') and (day<=30 and day>=22))
    or (month == ('июля' or 'июль') and (day>=1 and day <=20)):
        print('Ваш знак зодиака: Близнецы') 
    elif (month == ('июля' or 'июль') and (day<=31 and day>=21))
    or ( month == ('августа' or 'август') and (day>=1 and day <=10)):
        print('Ваш знак зодиака: Рак')
    elif (month == ('августа' or 'август') and (day<=31 and day>=11))
    or (month == ('сентября' or 'сентябрь') and (day>=1 and day <=16)):
        print('Ваш знак зодиака: Лев') 
    elif (month == ('сентября' or 'сентябрь') and (day<=31 and day>=17))
    or ( month == ('октября' or 'октябрь') and (day>=1 and day <=30)):
        print('Ваш знак зодиака: Дева')    
    elif (month == ('октября' or 'октябрь') and (day<=31 and day>=31))
    or (month == ('ноября' or 'ноябрь') and (day>=1 and day <=23)):
        print('Ваш знак зодиака: Весы') 
    elif (month == ('ноября' or 'ноябрь') and (day>=24 and day <=29)):
        print('Ваш знак зодиака: Скорпион')
    elif (month == ('ноября' or 'ноябрь') and day==30)
    or (month == ('декабря' or 'декабрь') and (day>=1 and day <=17)):
        print('Ваш знак зодиака: Змееносец') 
    elif (month == ('декабря' or 'декабрь') and (day<=31 and day>=18))
    or ( month == ('январь' or 'января') and (day>=1 and day <=20)):
        print('Ваш знак зодиака: Стрелец') 
    else:
        print('введите правильное название месяца!')
else:
    print('день находится вне диапазона для месяца, укажите от 1 до 31')   
````

## Задание 4

## Решение 4

````python
width = int(input('Ширину в (см) = '))
length = int(input('Длина в (см) = '))
height = int(input('Высота в (см) = '))
if width <= 15 and length <= 15 and height <= 15:
    print('Коробка №1')
# elif (width > 15 and width < 50) and (length > 15 and length < 50) and (height > 15 and height < 50):
elif (15 < width < 50) and (15 < length < 50) and (15 < height < 50):
    print('Коробка №2')
elif (150 > width) and (200 < length) and (150 > height):
    print('Ищите упаковку для лыж')
else:
    print('Стандартная коробка №3')
````

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

1. ```number = 123456```

Результат:  
`Несчастливый билет`
2. ```number = 123321```

## Решение 5

````python
a,b,c,d,x,y = input('Введите 6 значный код: \n')
number1 = int(a)+int(b)+int(c)
number2 = int(d)+int(x)+int(y)
if number1 == number2:
    print('Счастливый билет')
else:
    print('Несчастливый билет')
````

## Задание 6 (необязательное)

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус, S = 3,14 × r2
- если треугольник – длины трех его сторон; S = (a+b+c)/2
- если прямоугольник – длины двух его сторон. S = a × b

Пример работы программы:
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

## Решение 6

````python
user_input = input('Введите тип фигуры (Круг, Треугольник или Прямоугольник):\n')
if user_input == ('круг'or'Круг'):
    r = int(input('Введите радиус круга:\n'))
    print('Площадь круга:', (3.14*(r*r)))
if user_input ==('Треугольник'or 'треугольник'):
    a = int(input('Введите длину стороны A:\n'))
    b = int(input('Введите длину стороны B:\n'))
    c = int(input('Введите длину стороны C:\n'))
    print('Площадь треугольника:',(a+b+c)/2)
if user_input == ('прямоугольник' or 'Прямоугольник'):
    a = int(input('Введите длину стороны A:\n'))
    b = int(input('Введите длину стороны B:\n'))
    print('Площадь прямоугольника', a * b)
    
````
